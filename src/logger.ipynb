```python
# read from serial and dump to csv file with timestamp

import serial
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
com = 'COM7'
baud = 921600
ser = serial.Serial(com, baud, timeout=1)
```

```python
def read_serial(ser):
    line = ser.readline()
    line = line.decode('utf-8')
    return line
```

```python
filename = 'perfectsupersoft0_5mm_2.csv'

with open(filename, 'w') as f:
    f.write('time, data\n')

raw = []

while True:
    try:
        line = read_serial(ser)
        if line:
            print(line)
            raw.append(line)
            with open(filename, 'a') as f:
                # hh:mm:ss.ms data
                f.write(f'{time.strftime("%H:%M:%S")}.{int(time.time()*1000)%1000},{line}')
    except KeyboardInterrupt:
        break
    except:
        pass
```
