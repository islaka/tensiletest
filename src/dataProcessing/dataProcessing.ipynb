```python
# import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# variables
steps_per_mm = 100.0
```

```python
def process_data(testData, steps_per_mm):
    armed = testData[testData['data'] == 'Armed'].index[0]
    disarmed = testData[testData['data'] == 'Disarmed'].index[0]
    testData = testData[armed+1:disarmed]
    testData['data'] = testData['data'].astype(float)
    testData['data (N)'] = testData['data']*5
    testData.reset_index(drop=True, inplace=True)
    testData['steps'] = np.arange(0, len(testData)*10, 10)
    testData['displacement'] = testData['steps']/steps_per_mm
    return testData

def plot_data(testData, title):
    max_value = round(testData['data (N)'].max(),2)
    max_index = testData['data (N)'].idxmax()
    plt.plot(testData['displacement'], testData['data (N)'])
    plt.text(testData['displacement'][max_index], max_value, f'UTS: {max_value}', fontsize=8)
    plt.xlabel('Displacement (mm)')
    plt.ylabel('Force (N)')
    plt.title(f'{title} Tensile Test Line Plot')
    plt.show()
    plt.scatter(testData['displacement'], testData['data (N)'])
    # print text at testData max value
    plt.text(testData['displacement'][max_index], max_value, f'UTS: {max_value}', fontsize=8)
    plt.xlabel('Displacement (mm)')
    plt.ylabel('Force (N)')
    plt.title('Tensile Test Scatter Plot')
    plt.show()
```

```python
# import data
greenLine = pd.read_csv('greenline2.csv',header=0)
# greenLine.head()
ps05 = pd.read_csv('perfectsupersoft0_5mm_3.csv',header=0)
# ps05.head()
ps026 = pd.read_csv('perfectsupersoft0_26mm_1.csv',header=0)
# ps026.head()
kite = pd.read_csv('kite.csv',header=0)
guitarString = pd.read_csv('guitar_string.csv',header=0)
```

# 1. Green Line

```python
# preprocessing
greenLine_processed = process_data(greenLine, steps_per_mm)
greenLine_processed.describe()
```

```python
# data plot
plot_data(greenLine_processed, 'Green Line')
```

## 2. PS05

```python
# preprocessing
ps05_processed = process_data(ps05, steps_per_mm)
```

```python
# data plot
plot_data(ps05_processed, 'Perfect Super Soft 0.5mm')
```

## 3. PS026

```python
# preprocessing
ps026_processed = process_data(ps026, steps_per_mm)
```

```python
# data plot
plot_data(ps026_processed, 'Perfect Super Soft 0.26mm')
```

## 4. Kite

```python
# preprocessing
kite_processed = process_data(kite, steps_per_mm)
```

```python
# data plot
plot_data(kite_processed, 'Kite')
```

## 5. Guitar String

```python
# preprocessing
guitarString_processed = process_data(guitarString, steps_per_mm)
```

```python
# data plot
plot_data(guitarString_processed, 'Guitar String')
```
