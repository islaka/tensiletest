# 1. Green Line

```python
# import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# variables
steps_per_mm = 100.0
```

```python
# import data
greenLine = pd.read_csv('greenline2.csv',header=0)
greenLine.head()
ps05 = pd.read_csv('perfectsupersoft0_5mm_1.csv',header=0)
ps05.head()
ps026 = pd.read_csv('perfectsupersoft0_26mm_0.csv',header=0)
ps026.head()
```

```python
# preprocessing
testData = greenLine # change the data to be analyzed here
armed = testData[testData['data'] == 'Armed'].index[0]
disarmed = testData[testData['data'] == 'Disarmed'].index[0]
testData = testData[armed+1:disarmed]
# convert data from string to float
testData['data'] = testData['data'].astype(float)
# divide the data by 2 because the line was folded in half
testData['data (N)'] = testData['data']*5
# reset index
testData.reset_index(drop=True, inplace=True)
# create column steps, knowing that each line is an increment of 10 steps
testData['steps'] = np.arange(0, len(testData)*10, 10)
# create displacement column in mm
testData['displacement'] = testData['steps']/steps_per_mm
testData.head()
```

```python
# data plot
plt.plot(testData['displacement'], testData['data (N)'])
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (N)')
plt.title('Green Line Tensile Test Line Plot')
plt.show()
plt.scatter(testData['displacement'], testData['data (N)'])
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (N)')
plt.title('Green Line Tensile Test Scatter Plot')
plt.show()
```
