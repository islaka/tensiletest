```python
# import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# variables
steps_per_mm = 100.0
```

```python
def process_data(testData, steps_per_mm):
    armed = testData[testData['data'] == 'Armed'].index[0]
    disarmed = testData[testData['data'] == 'Disarmed'].index[0]
    testData = testData[armed+1:disarmed]
    testData['data'] = testData['data'].astype(float)
    testData['data (N)'] = testData['data']*5
    testData.reset_index(drop=True, inplace=True)
    testData['steps'] = np.arange(0, len(testData)*10, 10)
    testData['displacement'] = testData['steps']/steps_per_mm
    return testData

def plot_data(testData):
    plt.plot(testData['displacement'], testData['data (N)'])
    plt.xlabel('Displacement (mm)')
    plt.ylabel('Force (N)')
    plt.title('Tensile Test Line Plot')
    plt.show()
    plt.scatter(testData['displacement'], testData['data (N)'])
    plt.xlabel('Displacement (mm)')
    plt.ylabel('Force (N)')
    plt.title('Tensile Test Scatter Plot')
    plt.show()
```

# 1. Green Line

```python
# import data
greenLine = pd.read_csv('greenline2.csv',header=0)
greenLine.head()
ps05 = pd.read_csv('perfectsupersoft0_5mm_1.csv',header=0)
ps05.head()
ps026 = pd.read_csv('perfectsupersoft0_26mm_0.csv',header=0)
ps026.head()
```

```python
# preprocessing
greenLine_processed = process_data(greenLine, steps_per_mm)
```

```python
# data plot
plot_data(greenLine_processed)
```

## 2. PS05

```python
# preprocessing
ps05_processed = process_data(ps05, steps_per_mm)
```

```python
# data plot
plot_data(ps05_processed)
```

## 3. PS026

```python
# preprocessing
ps026_processed = process_data(ps026, steps_per_mm)
```

```python
# data plot
plot_data(ps026_processed)
```
